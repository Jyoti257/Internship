1) Write a python program to display IMDB’s Top rated 100 Indian movies’ data https://www.imdb.com/list/ls056092300/ (i.e. name, rating, year ofrelease) and make data frame.

```python
import re
url = 'https://www.imdb.com/list/ls091520106/'
response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')
movies = []
for movie in soup.find_all('div', class_='lister-item-content'):
    name = movie.find('h3', class_='lister-item-header').a.text.strip()
    rating = movie.find('span', class_='ipl-rating-star__rating').text.strip()
    year_text = movie.find('span', class_='lister-item-year').text.strip('()')
    year = re.findall(r'\d+', year_text)[0]
    movies.append({'Name': name, 'Rating': rating, 'Year': year})
df = pd.DataFrame(movies)
print(df)
```

4. Write a python program to scrape details of all the posts from https://www.patreon.com/coreyms .Scrape the 
heading, date, content and the likes for the video from the link for the youtube video from the post. 

```python
url = 'https://www.patreon.com/coreyms'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

posts = soup.find_all('div', class_='post')

for post in posts:
  heading = post.find('h3').text
  date = post.find('time')['datetime']
  content = post.find('div', class_='post__content').text
  youtube_link = post.find('a', class_='post__youtube-link')['href']
  likes = post.find('span', class_='post__likes').text

  print(f'Heading: {heading}')
  print(f'Date: {date}')
  print(f'Content: {content}')
  print(f'YouTube Link: {youtube_link}')
  print(f'Likes: {likes}')
  print('\n')
```

6) Write a python program to scrape first 10 product details which include product name , price , Image URL from 
https://www.bewakoof.com/bestseller?sort=popular . 

```python
response = requests.get("https://www.bewakoof.com/bestseller?sort=popular")

soup = BeautifulSoup(response.text, 'html.parser')

product_cards = soup.find_all('div', class_='productCardBox')

for card in product_cards[:10]:
  
    product_name_element = card.find('p', class_='name')
    product_name = product_name_element.text.strip() if product_name_element else "N/A"
    
 
    product_price_element = card.find('span', class_='price')
    product_price = product_price_element.text.strip() if product_price_element else "N/A"
    

    image_url_element = card.find('img')
    image_url = image_url_element['src'] if image_url_element else "N/A"
    
    print("Product Name:", product_name)
    print("Price:", product_price)
    print("Image URL:", image_url)
    print("\n")
```

7. Please visit https://www.cnbc.com/world/?region=world and scrap- 
a) headings 
b) date 
c) News link

```python
url = "https://www.cnbc.com/world/?region=world"
response = requests.get(url)
```

```python
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')

articles = soup.find_all('div', class_='Card-titleContainer')

for article in articles:
  
    heading_element = article.find('a')
    heading = heading_element.text.strip() if heading_element else "N/A"

    date_element = article.find('span', class_='Card-date')
    date = date_element.text.strip() if date_element else "N/A"
    
    news_link_element = article.find('a')
    news_link = news_link_element['href'] if news_link_element else "N/A"
    
    print("Heading:", heading)
    print("Date:", date)
    print("News Link:", news_link)
    print("\n")
```

```python
url = "https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/"
response = requests.get(url)
print(response.text)
```

```python
url = "https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/"
response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')

panels = soup.find_all('div', class_='panel')

for panel in panels:
    
    title = panel.find('h2', class_='article-title').text.strip()
    
    date = panel.find('p', class_='article-date').text.strip()
    
    authors = panel.find('p', class_='article-authors').text.strip()
    
    print("Paper Title:", title)
    print("Date:", date)
    print("Authors:", authors)
    print("\n")
```
